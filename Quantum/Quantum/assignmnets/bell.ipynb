```python
#prepared by:
#Youssef Hamdy Abdelmoneim Attallah 46-5543
#Omar Ahmed Altantawy 46-8626
import numpy as np
from qiskit import *
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

simulator = AerSimulator()

q = QuantumRegister(2)

def e1(quantum_circuit):
    circuit = quantum_circuit.copy()
    circuit.s(0)
    circuit.h(0)
    circuit.t(0)
    circuit.h(0)
    circuit.measure([0, 1], [0, 1])
    circuit1 = transpile(circuit, simulator)
    job = simulator.run(circuit1, shots=1000)
    counts = job.result().get_counts(circuit1)
    print(counts)
    return ret_e(counts)


def e2(quantum_circuit):
    circuit = quantum_circuit.copy()
    circuit.s(0)
    circuit.h(0)
    circuit.tdg(0)
    circuit.h(0)
    circuit.measure([0, 1], [0, 1])
    circuit1 = transpile(circuit, simulator)
    job = simulator.run(circuit1, shots=1000)
    counts = job.result().get_counts(circuit1)
    print(counts)
    return ret_e(counts)

def e3(quantum_circuit):
    circuit = quantum_circuit.copy()
    circuit.s(0)
    circuit.h(0)
    circuit.tdg(0)
    circuit.h(0)
    circuit.h(1)
    circuit.measure([0, 1], [0, 1])
    circuit1 = transpile(circuit, simulator)
    job = simulator.run(circuit1, shots=1000)
    counts = job.result().get_counts(circuit1)
    print(counts)
    return ret_e(counts)

def e4(quantum_circuit):
    circuit = quantum_circuit.copy()
    circuit.s(0)
    circuit.h(0)
    circuit.t(0)
    circuit.h(0)
    circuit.h(1)
    circuit.measure([0, 1], [0, 1])
    circuit1 = transpile(circuit, simulator)
    job = simulator.run(circuit1, shots=1000)
    counts = job.result().get_counts(circuit1)
    print(counts)
    return ret_e(counts)

def ret_e(counts):
    try:
        num1 = counts.get('00')/1000
    except:
        num1 = 0

    try:
        num2 = counts.get('01')/1000
    except:
        num2 = 0

    try:
        num3 = counts.get('10')/1000
    except:
        num3 = 0

    try:
        num4 = counts.get('11')/1000
    except:
        num4 = 0

    return num1 - num2 - num3 + num4


def ret_S_param(circuit):
    return e1(circuit) + e2(circuit) + e4(circuit) - e3(circuit)
#qcircuit.measure_all()
#qcircuit.draw()
```

```python
def maximally_entangled_circuit():
    circuit = QuantumCircuit(2, 2)
    circuit.h(0)
    circuit.cx(0, 1)
    return circuit

def partially_entangled_circuit():
    circuit = QuantumCircuit(2, 2)
    #circuit.h(0)
    circuit.initialize([np.sqrt(0.48), np.sqrt(0.02), np.sqrt(0.02), np.sqrt(0.48)], circuit.qubits)
    return circuit

def nonentangled_circuit():
    circuit = QuantumCircuit(2, 2)
    return circuit


print("Maximally entangled circuit:" + str(ret_S_param(maximally_entangled_circuit())))
print("Partially entangled circuit:" + str(ret_S_param(partially_entangled_circuit())))
print("Nonentangled circuit:" + str(ret_S_param(nonentangled_circuit())))
```
